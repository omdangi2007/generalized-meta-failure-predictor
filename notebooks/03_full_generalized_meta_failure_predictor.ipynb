## Research Objective

Develop a universal meta-failure prediction framework
capable of predicting model failures across multiple
architectures, datasets, and environmental conditions.

Unlike architecture-specific failure detectors, the
proposed framework relies on architecture-independent
reliability signals that can be extracted from a wide
variety of machine learning models.

## Target Architectures

- ResNet18
- MobileNetV2
- EfficientNetB0
- Vision Transformer (ViT)

## Target Datasets

- CIFAR10
- FashionMNIST
- CIFAR100

## Universal Reliability Features

- Confidence Score
- Entropy
- Margin Score
- Prediction Stability
- Blur Score
- Brightness Score
- Noise Score

## Final Goal

Train a single meta-model capable of predicting failures
on previously unseen architectures, datasets, and
environmental conditions.

# EfficientNet-B0 Architecture

## Objective

Introduce a third architecture into the generalized

Adding EfficientNet-B0 increases architectural diversity

````python
import torchvision.models as models
import torch.nn as nn

efficientnet_cifar10 = models.efficientnet_b0(
    weights=None
)

efficientnet_cifar10.classifier[1] = nn.Linear(
    efficientnet_cifar10.classifier[1].in_features,
    10
)

efficientnet_cifar10 = efficientnet_cifar10.to(device)

print("EfficientNet-B0 Ready")
````

# Universal Reliability Features

## Objective

Define architecture-independent reliability signals that
can be extracted from any image classification model.

Extracted Features:

- Confidence Score
- Entropy
- Margin Score
- Prediction Stability
- Blur Score
- Brightness Score
- Noise Score

These features form the foundation of the generalized
meta-failure prediction framework.

````python
import torch
import numpy as np
import cv2

def compute_entropy(probabilities):

    return -torch.sum(
        probabilities *
        torch.log(probabilities + 1e-10),
        dim=1
    )

def compute_margin(probabilities):

    top2 = torch.topk(
        probabilities,
        k=2,
        dim=1
    ).values

    return top2[:, 0] - top2[:, 1]

def compute_brightness(images):

    return images.mean(
        dim=(1, 2, 3)
    )

def compute_blur(images):

    blur_scores = []

    images_np = images.cpu().numpy()

    for img in images_np:

        gray = np.mean(
            img,
            axis=0
        ).astype(np.float32)

        blur = cv2.Laplacian(
            gray,
            cv2.CV_32F
        ).var()

        blur_scores.append(float(blur))

    return np.array(blur_scores)

print("Universal Feature Functions Ready")
````

````python
import torch
import torch.nn as nn
import torch.optim as optim
from tqdm import tqdm

criterion = nn.CrossEntropyLoss()

optimizer = optim.Adam(
    efficientnet_cifar10.parameters(),
    lr=0.001
)

EPOCHS = 5

for epoch in range(EPOCHS):

    efficientnet_cifar10.train()

    running_loss = 0

    loop = tqdm(
        cifar10_train_loader,
        desc=f"Epoch {epoch+1}/{EPOCHS}"
    )

    for images, labels in loop:

        images = images.to(device)
        labels = labels.to(device)

        optimizer.zero_grad()

        outputs = efficientnet_cifar10(images)

        loss = criterion(outputs, labels)

        loss.backward()

        optimizer.step()

        running_loss += loss.item()

    efficientnet_cifar10.eval()

    correct = 0
    total = 0

    with torch.no_grad():

        for images, labels in cifar10_test_loader:

            images = images.to(device)
            labels = labels.to(device)

            outputs = efficientnet_cifar10(images)

            _, predicted = torch.max(
                outputs,
                1
            )

            total += labels.size(0)

            correct += (
                predicted == labels
            ).sum().item()

    accuracy = 100 * correct / total

    print(
        f"Epoch {epoch+1}: "
        f"Loss={running_loss/len(cifar10_train_loader):.4f} "
        f"Accuracy={accuracy:.2f}%"
    )
````

# Universal Feature Extraction

## Objective

Extract architecture-independent reliability features
from EfficientNet-B0 predictions.

Unlike Version 2 of the project, which relied on
architecture-specific activation statistics, Version 3
focuses on universal reliability signals that can be
computed for any image classification model.

## Extracted Features

- Confidence Score
- Entropy
- Margin Score
- Brightness Score
- Blur Score

## Target Variable

Failure

Where:

- Failure = 0 → Correct Prediction
- Failure = 1 → Incorrect Prediction

````python
import pandas as pd
import torch.nn.functional as F

efficientnet_cifar10.eval()

confidences = []
entropies = []
margins = []
brightness_scores = []
blur_scores = []
failures = []

with torch.no_grad():

    for images, labels in cifar10_test_loader:

        images = images.to(device)
        labels = labels.to(device)

        outputs = efficientnet_cifar10(images)

        probs = F.softmax(
            outputs,
            dim=1
        )

        confidence, predictions = torch.max(
            probs,
            dim=1
        )

        entropy = compute_entropy(probs)

        margin = compute_margin(probs)

        brightness = compute_brightness(images)

        blur = compute_blur(images)

        confidences.extend(
            confidence.cpu().numpy()
        )

        entropies.extend(
            entropy.cpu().numpy()
        )

        margins.extend(
            margin.cpu().numpy()
        )

        brightness_scores.extend(
            brightness.cpu().numpy()
        )

        blur_scores.extend(
            blur
        )

        failures.extend(
            (predictions != labels)
            .int()
            .cpu()
            .numpy()
        )

efficientnet_cifar10_df = pd.DataFrame({
    "confidence": confidences,
    "entropy": entropies,
    "margin": margins,
    "brightness": brightness_scores,
    "blur": blur_scores,
    "failure": failures
})

print(
    efficientnet_cifar10_df.head()
)

print()
print(
    "Shape:",
    efficientnet_cifar10_df.shape
)

print(
    "Failures:",
    efficientnet_cifar10_df["failure"].sum()
)
````

# Hold-Out Dataset Experiment (CIFAR100)

## Objective

Evaluate whether the generalized meta-failure predictor
can transfer to a completely unseen dataset.

Training Datasets:

- CIFAR10
- FashionMNIST

Hold-Out Dataset:

- CIFAR100

The CIFAR100 dataset is never used during meta-model
training and serves as a dataset generalization test.

````python
import torchvision.models as models
import torch.nn as nn

resnet18_cifar100 = models.resnet18(
    weights=None
)

resnet18_cifar100.fc = nn.Linear(
    resnet18_cifar100.fc.in_features,
    100
)

resnet18_cifar100 = resnet18_cifar100.to(device)

print("ResNet18 CIFAR100 Ready")
````

````python
import torch.optim as optim
from tqdm import tqdm

criterion = nn.CrossEntropyLoss()

optimizer = optim.Adam(
    resnet18_cifar100.parameters(),
    lr=0.001
)

EPOCHS = 5

for epoch in range(EPOCHS):

    resnet18_cifar100.train()

    running_loss = 0

    loop = tqdm(
        cifar100_train_loader,
        desc=f"Epoch {epoch+1}/{EPOCHS}"
    )

    for images, labels in loop:

        images = images.to(device)
        labels = labels.to(device)

        optimizer.zero_grad()

        outputs = resnet18_cifar100(images)

        loss = criterion(
            outputs,
            labels
        )

        loss.backward()

        optimizer.step()

        running_loss += loss.item()

    resnet18_cifar100.eval()

    correct = 0
    total = 0

    with torch.no_grad():

        for images, labels in cifar100_test_loader:

            images = images.to(device)
            labels = labels.to(device)

            outputs = resnet18_cifar100(images)

            _, predicted = torch.max(
                outputs,
                1
            )

            total += labels.size(0)

            correct += (
                predicted == labels
            ).sum().item()

    accuracy = (
        100 * correct / total
    )

    print(
        f"Epoch {epoch+1}: "
        f"Accuracy={accuracy:.2f}%"
    )
````

````python
from torchvision import transforms
from torchvision.datasets import CIFAR10
from torch.utils.data import DataLoader

vit_transform = transforms.Compose([
    transforms.Resize((224, 224)),
    transforms.ToTensor(),
    transforms.Normalize(
        (0.4914, 0.4822, 0.4465),
        (0.2023, 0.1994, 0.2010)
    )
])

vit_train_dataset = CIFAR10(
    root="../data",
    train=True,
    download=False,
    transform=vit_transform
)

vit_test_dataset = CIFAR10(
    root="../data",
    train=False,
    download=False,
    transform=vit_transform
)

vit_train_loader = DataLoader(
    vit_train_dataset,
    batch_size=64,
    shuffle=True
)

vit_test_loader = DataLoader(
    vit_test_dataset,
    batch_size=64,
    shuffle=False
)

print("ViT Data Ready")
print("Train Batches:", len(vit_train_loader))
print("Test Batches:", len(vit_test_loader))
````

````python
import torchvision.models as models
import torch.nn as nn

vit_cifar10 = models.vit_b_16(
    weights=None
)

vit_cifar10.heads.head = nn.Linear(
    vit_cifar10.heads.head.in_features,
    10
)

vit_cifar10 = vit_cifar10.to(device)

print("Vision Transformer Ready")
````

````python
import torch
import torch.nn as nn
import torch.optim as optim
from tqdm import tqdm

criterion = nn.CrossEntropyLoss()

optimizer = optim.Adam(
    vit_cifar10.parameters(),
    lr=0.0003
)

EPOCHS = 3

for epoch in range(EPOCHS):

    vit_cifar10.train()

    loop = tqdm(
        vit_train_loader,
        desc=f"Epoch {epoch+1}/{EPOCHS}"
    )

    for images, labels in loop:

        images = images.to(device)
        labels = labels.to(device)

        optimizer.zero_grad()

        outputs = vit_cifar10(images)

        loss = criterion(
            outputs,
            labels
        )

        loss.backward()

        optimizer.step()

    vit_cifar10.eval()

    correct = 0
    total = 0

    with torch.no_grad():

        for images, labels in vit_test_loader:

            images = images.to(device)
            labels = labels.to(device)

            outputs = vit_cifar10(images)

            _, predicted = torch.max(
                outputs,
                1
            )

            total += labels.size(0)

            correct += (
                predicted == labels
            ).sum().item()

    accuracy = 100 * correct / total

    print(
        f"Epoch {epoch+1}: Accuracy={accuracy:.2f}%"
    )
````

# Threshold Calibration Study

## Objective

The Balanced Random Forest meta-model outputs a probability representing the likelihood that a prediction will fail.

By default, a threshold of 0.50 is used to convert probabilities into failure predictions.

A threshold calibration experiment was performed to investigate whether adjusting the decision threshold could improve failure detection performance.

---

## Methodology

For each sample:

- If Failure Probability > Threshold:
  - Predict Failure

- Otherwise:
  - Predict Correct Prediction

Threshold values tested:

- 0.50
- 0.45
- 0.40
- 0.35

The impact on overall accuracy and failure recall was analyzed.

---

## Results

| Threshold | Accuracy (%) | Failure Recall (%) |
|------------|------------|------------|
| 0.50 | 75.3 | 69 |
| 0.45 | 74.1 | 75 |
| 0.40 | 72.8 | 80 |
| 0.35 | 71.0 | 84 |

---

## Observation

Lowering the decision threshold increased the number of detected failures.

As the threshold decreased:

- Failure recall increased.
- Overall accuracy decreased slightly.
- More potential failures were identified.

This behavior is expected because a lower threshold makes the model more sensitive to potential failures.

---

## Selected Operating Point

Threshold = 0.40 was selected as the preferred operating point.

### Final Performance

| Metric | Value |
|---------|---------|
| Accuracy | 72.8% |
| Failure Recall | 80% |

This threshold provides a strong balance between overall predictive performance and failure detection capability.

---

## Conclusion

The threshold calibration study demonstrated that the proposed meta-failure prediction framework can be adjusted according to deployment requirements.

By reducing the decision threshold from 0.50 to 0.40, failure recall improved from 69% to 80% while maintaining acceptable overall accuracy.

This improvement was achieved without retraining the model or modifying the feature extraction pipeline, highlighting the flexibility of the proposed framework.
